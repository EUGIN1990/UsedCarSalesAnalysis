## 6.1 Handling Missing Values

**1).EngV -----> missing values**

```python
car_data['engV'].isnull().sum() # checking the sum of null values in the column
```

```python
car_data['engV'].agg(['median','mean','min','max'])
```

```python
car_data['engV']=car_data['engV'].fillna(car_data['engV'].median())
```

**Verifying**

```python
car_data['engV'].isnull().sum() # checking all the missing values are filled
```

```python
car_data['engV'].describe() #The count shows that all the value are filled with median
```

**2).drive ----> missing values**

```python
#checking the null values

car_data['drive'].isnull().sum()
```

```python
# checking the value count

car_data['drive'].value_counts()
```

```python
#its a catagorical value so replace the missing value with mode

car_data['drive'].mode()[0]
```

```python
car_data['drive']=car_data['drive'].fillna(car_data['drive'].mode()[0])
```

**Verifying**

```python
car_data['drive'].isnull().sum()
```

```python
car_data['drive'].describe()
```

## 6.2 Handling Outlier

**1).EngV -----> maximum value ---> outlier**

```python
car_data['engV'].plot.box()
```

```python
#checking the upper_acceptable value 

upper_val=car_data['engV'].quantile(0.98)
upper_val
```

```python
#checking the lower_acceptable value 
lower_val=car_data['engV'].quantile(0.02)
lower_val
```

```python
#adjustin the outlier to lower and upper accepatable values

car_data['engV']=car_data['engV'].clip(upper=upper_val)
car_data['engV']=car_data['engV'].clip(lower=lower_val)
```

**Verify**

```python
car_data['engV'].plot.box()
```

**2).Milage ---> check outlier**

```python
car_data['mileage'].describe() 
```

```python
#checking the outlier using box plot

car_data['mileage'].plot.box()
```

```python
upper_val=car_data['mileage'].quantile(0.98) # checking the upper adjustable value
upper_val
```

```python
car_data['mileage']=car_data['mileage'].clip(upper=upper_val) # adjusting the outlier using upper adjustable value
```

**Verify**

```python
car_data['mileage'].plot.box()
```

**1. List the top 5 car brands by sales,and how many units did they sell and  how do the sales figures compare across the top 5 brands?**

```python
brand_sales = car_data['car'].value_counts()
top_5_brands = brand_sales.head(5)
plt.figure(figsize=(10, 6))
bars = plt.bar(top_5_brands.index, top_5_brands.values, color=['blue', 'green', 'red', 'purple', 'orange'])


plt.title('Top 5 Car Brands by Sales')
plt.xlabel('Brand')
plt.ylabel('Number of Sales')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom')

plt.show()
```

**Observation**

The bar graph visually represents the sales distribution among the top 5 car brands. Volkswagen leads with the highest number of sales, closely followed by Mercedes-Benz. 

There is a noticeable drop in sales figures from Mercedes-Benz to BMW, indicating that the top two brands dominate the market significantly. 

Toyota and VAZ have similar sales figures but are clearly behind the top three brands. This suggests that while Volkswagen and Mercedes-Benz are the most popular among consumers, there is a competitive gap that BMW, Toyota, and VAZ have yet to bridge.
